```python
import tensorflow as tf
import pathlib
import random
import os

# Set random seed for reproducibility
SEED = 42
random.seed(SEED)
tf.random.set_seed(SEED)

def load_and_preprocess_image(image_path, is_sketch=False):
    # Read the image file
    image = tf.io.read_file(image_path)

    # Decode the image (use decode_jpeg for JPEG images)
    image = tf.image.decode_jpeg(image, channels=3)

    # Ensure the image has a shape
    image.set_shape([None, None, 3])

    # Convert to float32 and normalize to [-1, 1]
    image = tf.cast(image, tf.float32)
    image = (image / 127.5) - 1

    # If it's a sketch, convert to grayscale
    if is_sketch:
        image = tf.image.rgb_to_grayscale(image)

    return image

def resize_image(image, height, width):
    return tf.image.resize(image, [height, width], method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)

def random_crop(image, height, width):
    cropped_image = tf.image.random_crop(image, size=[height, width, tf.shape(image)[-1]])
    return cropped_image

def normalize(image):
    return (image / 127.5) - 1

@tf.function
def random_jitter(sketch, real):
    # Resize to 286x286
    sketch = resize_image(sketch, 286, 286)
    real = resize_image(real, 286, 286)

    # Ensure the sketch also has 3 channels before stacking
    sketch = tf.tile(sketch, [1, 1, 3]) # Replicate the grayscale channel 3 times

    # Random crop back to 256x256
    # Remove stacking of tensors
    # stacked = tf.stack([sketch, real], axis=0)
    # cropped = random_crop(stacked, 256, 256)
    # sketch, real = cropped[0], cropped[1]
    sketch = random_crop(sketch, 256, 256)
    real = random_crop(real, 256, 256)

    # Random mirroring
    if tf.random.uniform(()) > 0.5:
        sketch = tf.image.flip_left_right(sketch)
        real = tf.image.flip_left_right(real)

    return sketch, real

def load_image_pair(sketch_path, real_path):
    sketch = load_and_preprocess_image(sketch_path, is_sketch=True)
    real = load_and_preprocess_image(real_path, is_sketch=False)

    return sketch, real

def create_dataset(sketch_dir, real_dir, batch_size=1, shuffle_buffer_size=1000):
    # Get sorted lists of sketch and real image files
    sketch_files = sorted([str(p) for p in pathlib.Path(sketch_dir).glob('*.jpg')])
    real_files = sorted([str(p) for p in pathlib.Path(real_dir).glob('*.jpg')])

    if not sketch_files or not real_files:
        raise ValueError(f"No JPEG images found in {sketch_dir} or {real_dir}")

    # Create tf.data.Dataset
    dataset = tf.data.Dataset.from_tensor_slices((sketch_files, real_files))

    # Load and preprocess image pairs
    dataset = dataset.map(lambda s, r: tf.py_function(load_image_pair, [s, r], [tf.float32, tf.float32]),
                          num_parallel_calls=tf.data.AUTOTUNE)

    # Set the shapes explicitly
    dataset = dataset.map(lambda s, r: (tf.ensure_shape(s, [None, None, 1]), tf.ensure_shape(r, [None, None, 3])))

    # Apply random jitter and mirroring
    dataset = dataset.map(random_jitter, num_parallel_calls=tf.data.AUTOTUNE)

    # Shuffle, batch, and prefetch
    dataset = dataset.shuffle(shuffle_buffer_size, seed=SEED)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)

    return dataset

# Example usage

SKETCH_DIR = '/content/drive/MyDrive/New Dataset for pix2pix/Train 2/Sketches'
REAL_DIR = '/content/drive/MyDrive/New Dataset for pix2pix/Train 2/Photos'




BATCH_SIZE = 16

# Create train and test datasets
train_dataset = create_dataset(SKETCH_DIR, REAL_DIR, batch_size=BATCH_SIZE)
test_dataset = create_dataset(SKETCH_DIR, REAL_DIR, batch_size=1)  # No augmentation for test set

# Function to visualize some samples
def visualize_samples(dataset, num_samples=3):
    import matplotlib.pyplot as plt

    sample_sketches, sample_reals = next(iter(dataset.take(num_samples)))

    plt.figure(figsize=(12, 4*num_samples))
    for i in range(num_samples):
        plt.subplot(num_samples, 2, 2*i+1)
        plt.imshow(sample_sketches[i, ..., 0], cmap='gray')
        plt.title(f"Sketch {i+1}")
        plt.axis('off')

        plt.subplot(num_samples, 2, 2*i+2)
        plt.imshow((sample_reals[i] + 1) / 2)  # Denormalize for visualization
        plt.title(f"Real {i+1}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Visualize some samples
visualize_samples(train_dataset)

# Error handling: Try to fetch a batch to check if the dataset is working
try:
    for sketch_batch, real_batch in train_dataset.take(1):
        print("Dataset is working!")
        print(f"Sketch batch shape: {sketch_batch.shape}")
        print(f"Real batch shape: {real_batch.shape}")
except Exception as e:
    print(f"Error occurred while fetching a batch: {str(e)}")
    print("Please check your image directories and ensure the images are valid.")
```

```python
import tensorflow as tf
from tensorflow import keras

class InstanceNormalization(keras.layers.Layer):
    def __init__(self, epsilon=1e-5):
        super(InstanceNormalization, self).__init__()
        self.epsilon = epsilon

    def build(self, input_shape):
        self.scale = self.add_weight(
            name='scale',
            shape=input_shape[-1:],
            initializer='ones',
            trainable=True)

        self.offset = self.add_weight(
            name='offset',
            shape=input_shape[-1:],
            initializer='zeros',
            trainable=True)

    def call(self, x):
        mean, variance = tf.nn.moments(x, axes=[1, 2], keepdims=True)
        inv = tf.math.rsqrt(variance + self.epsilon)
        normalized = (x - mean) * inv
        return self.scale * normalized + self.offset

def downsample(filters, size, apply_norm=True):
    initializer = tf.random_normal_initializer(0., 0.02)
    result = keras.Sequential()
    result.add(keras.layers.Conv2D(filters, size, strides=2, padding='same',
                                   kernel_initializer=initializer, use_bias=False))
    if apply_norm:
        result.add(InstanceNormalization())
    result.add(keras.layers.LeakyReLU())
    return result

def upsample(filters, size, apply_dropout=False):
    initializer = tf.random_normal_initializer(0., 0.02)
    result = keras.Sequential()
    result.add(keras.layers.Conv2DTranspose(filters, size, strides=2, padding='same',
                                            kernel_initializer=initializer, use_bias=False))
    result.add(InstanceNormalization())
    if apply_dropout:
        result.add(keras.layers.Dropout(0.5))
    result.add(keras.layers.ReLU())
    return result

def Generator():
    inputs = keras.layers.Input(shape=[256, 256, 3])

    # Encoder (downsampling)
    down_stack = [
        downsample(64, 4, apply_norm=False),  # (128, 128, 64)
        downsample(128, 4),  # (64, 64, 128)
        downsample(256, 4),  # (32, 32, 256)
        downsample(512, 4),  # (16, 16, 512)
        downsample(512, 4),  # (8, 8, 512)
        downsample(512, 4),  # (4, 4, 512)
        downsample(512, 4),  # (2, 2, 512)
        downsample(512, 4),  # (1, 1, 512)
    ]

    # Decoder (upsampling)
    up_stack = [
        upsample(512, 4, apply_dropout=True),  # (2, 2, 1024)
        upsample(512, 4, apply_dropout=True),  # (4, 4, 1024)
        upsample(512, 4, apply_dropout=True),  # (8, 8, 1024)
        upsample(512, 4),  # (16, 16, 1024)
        upsample(256, 4),  # (32, 32, 512)
        upsample(128, 4),  # (64, 64, 256)
        upsample(64, 4),  # (128, 128, 128)
    ]

    initializer = tf.random_normal_initializer(0., 0.02)
    last = keras.layers.Conv2DTranspose(3, 4, strides=2, padding='same',
                                        kernel_initializer=initializer,
                                        activation='tanh')  # (256, 256, 3)

    x = inputs
    skips = []
    for down in down_stack:
        x = down(x)
        skips.append(x)

    skips = reversed(skips[:-1])

    for up, skip in zip(up_stack, skips):
        x = up(x)
        x = keras.layers.Concatenate()([x, skip])

    x = last(x)

    return keras.Model(inputs=inputs, outputs=x)

def Discriminator():
    initializer = tf.random_normal_initializer(0., 0.02)

    inp = keras.layers.Input(shape=[256, 256, 3], name='input_image')
    tar = keras.layers.Input(shape=[256, 256, 3], name='target_image')

    x = keras.layers.concatenate([inp, tar])  # (256, 256, channels*2)

    down1 = downsample(64, 4, False)(x)  # (128, 128, 64)
    down2 = downsample(128, 4)(down1)  # (64, 64, 128)
    down3 = downsample(256, 4)(down2)  # (32, 32, 256)

    zero_pad1 = keras.layers.ZeroPadding2D()(down3)  # (34, 34, 256)
    conv = keras.layers.Conv2D(512, 4, strides=1,
                               kernel_initializer=initializer,
                               use_bias=False)(zero_pad1)  # (31, 31, 512)

    norm1 = InstanceNormalization()(conv)
    leaky_relu = keras.layers.LeakyReLU()(norm1)

    zero_pad2 = keras.layers.ZeroPadding2D()(leaky_relu)  # (33, 33, 512)
    last = keras.layers.Conv2D(1, 4, strides=1,
                               kernel_initializer=initializer)(zero_pad2)  # (30, 30, 1)

    return keras.Model(inputs=[inp, tar], outputs=last)

class Pix2Pix(keras.Model):
    def __init__(self):
        super(Pix2Pix, self).__init__()
        self.generator = Generator()
        self.discriminator = Discriminator()
        self.lambda_value = 100

    def compile(self, gen_optimizer, disc_optimizer, loss_fn):
        super(Pix2Pix, self).compile()
        self.gen_optimizer = gen_optimizer
        self.disc_optimizer = disc_optimizer
        self.loss_fn = loss_fn

    def train_step(self, batch):
        sketch, real = batch

        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            gen_output = self.generator(sketch, training=True)

            disc_real_output = self.discriminator([sketch, real], training=True)
            disc_generated_output = self.discriminator([sketch, gen_output], training=True)

            gen_loss = self.generator_loss(disc_generated_output, gen_output, real)
            disc_loss = self.discriminator_loss(disc_real_output, disc_generated_output)

        generator_gradients = gen_tape.gradient(gen_loss, self.generator.trainable_variables)
        discriminator_gradients = disc_tape.gradient(disc_loss, self.discriminator.trainable_variables)

        self.gen_optimizer.apply_gradients(zip(generator_gradients, self.generator.trainable_variables))
        self.disc_optimizer.apply_gradients(zip(discriminator_gradients, self.discriminator.trainable_variables))

        return {"gen_loss": gen_loss, "disc_loss": disc_loss}

    def generator_loss(self, disc_generated_output, gen_output, target):
        gan_loss = self.loss_fn(tf.ones_like(disc_generated_output), disc_generated_output)
        l1_loss = tf.reduce_mean(tf.abs(target - gen_output))
        return gan_loss + self.lambda_value * l1_loss

    def discriminator_loss(self, disc_real_output, disc_generated_output):
        real_loss = self.loss_fn(tf.ones_like(disc_real_output), disc_real_output)
        generated_loss = self.loss_fn(tf.zeros_like(disc_generated_output), disc_generated_output)
        return real_loss + generated_loss
```

```python
EPOCHS = 250
loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=True)
gen_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
disc_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)

model = Pix2Pix()
model.compile(gen_optimizer=gen_optimizer, disc_optimizer=disc_optimizer, loss_fn=loss_fn)

history = model.fit(train_dataset, epochs=EPOCHS)
```

```python
import tensorflow as tf
import numpy as np

# Function to evaluate SSIM and PSNR for a test dataset
def evaluate_model(model, test_dataset):
    ssim_metric = tf.image.ssim  # SSIM function in TensorFlow
    psnr_metric = tf.image.psnr  # PSNR function in TensorFlow

    ssim_values = []
    psnr_values = []

    # Loop through the test dataset
    for sketch, real in test_dataset:
        # Generate image using the generator
        generated = model.generator(sketch, training=False)

        # Ensure the images are in the correct range [0, 1] for SSIM and PSNR calculations
        real = (real + 1) / 2.0
        generated = (generated + 1) / 2.0

        # Compute SSIM and PSNR
        ssim = ssim_metric(real, generated, max_val=1.0)
        psnr = psnr_metric(real, generated, max_val=1.0)

        # Store values
        ssim_values.extend(ssim.numpy())
        psnr_values.extend(psnr.numpy())

    # Compute and print the average SSIM and PSNR
    avg_ssim = np.mean(ssim_values)
    avg_psnr = np.mean(psnr_values)

    print(f"Average SSIM: {avg_ssim:.4f}")
    print(f"Average PSNR: {avg_psnr:.4f}")

    return avg_ssim, avg_psnr

# Example usage after training
avg_ssim, avg_psnr = evaluate_model(model, test_dataset)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from gfpgan import GFPGANer  # Ensure GFPGAN is installed and importable

# Load GFPGAN model (Ensure you have the correct path to your pretrained model)
gfpganer = GFPGANer(model_path='/content/experiments/pretrained_models/GFPGANv1.3.pth.1', upscale=2, arch='clean', channel_multiplier=2)

# Function to load the Pix2Pix model from Google Drive
def load_model_from_drive(model_path):
    loaded_model = tf.keras.models.load_model(model_path, compile=False)
    print(f"Model loaded from {model_path}")
    return loaded_model

# Function to preprocess a single image
def preprocess_single_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [256, 256])
    image = (image / 127.5) - 1
    return image

# Function to apply GFPGAN for post-processing
def apply_gfpgan(image):
    # Convert image to uint8 for GFPGAN processing
    image_uint8 = np.uint8((image * 255).numpy())  # Convert to uint8 [0, 255]

    # Ensure image is in BGR format for GFPGAN
    image_bgr = cv2.cvtColor(image_uint8, cv2.COLOR_RGB2BGR)

    # Apply GFPGAN for face restoration
    _, restored_image = gfpganer.enhance(image_bgr, has_aligned=False, only_center_face=False)

    # Convert restored image back to RGB and normalize to [0, 1]
    restored_image_rgb = cv2.cvtColor(restored_image, cv2.COLOR_BGR2RGB)
    restored_image_rgb = np.float32(restored_image_rgb) / 255.0
    return restored_image_rgb

# Function to generate color image from sketch and apply GFPGAN for enhancement
def generate_and_enhance_image(model, sketch_path):
    # Preprocess the input sketch
    sketch = preprocess_single_image(sketch_path)
    sketch = tf.expand_dims(sketch, 0)  # Add batch dimension

    # Generate the color image using the Pix2Pix model
    generated_image = model.generator(sketch, training=False)
    generated_image = generated_image[0]  # Remove batch dimension

    # Apply GFPGAN post-processing for face restoration
    final_output = apply_gfpgan(generated_image * 0.5 + 0.5)  # Scaling back to [0, 1]

    # Plot the input sketch, generated image, and GFPGAN-enhanced image
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.title("Input Sketch")
    plt.imshow(sketch[0] * 0.5 + 0.5)
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Generated Image")
    plt.imshow(generated_image * 0.5 + 0.5)
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("GFPGAN Enhanced Image")
    plt.imshow(final_output)
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    return final_output

# Load the Pix2Pix model from Google Drive
model_path = '/content/drive/MyDrive/pix2pix_facial_sketch_to_color_final.keras'
model = load_model_from_drive(model_path)

# Provide the path to your sketch image on Google Drive
sketch_path = '/content/drive/MyDrive/New Dataset for pix2pix/Val 2/Sketches/00.jpg'  # Replace with your actual sketch path

# Generate and enhance the image using Pix2Pix and GFPGAN
generated_and_enhanced_image = generate_and_enhance_image(model, sketch_path)
```
